```python
import json
text=movies['genres'][0]
genre_list = json.loads(text)

print(genre_list)
```

```python
def genres_split(text):
    list= json.loads(text)
    gen=[]
    for i in list:
        gen.append(i['name'])
    return gen
```

```python
movies['genres']=movies['genres'].apply(genres_split)
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.replace(","," "))
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.replace("  "," "))
```

```python
movies['overview'][0]
```

```python
movies['keywords'][0]
```

```python
def keyword_split(text):
    list= json.loads(text)
    key=[]
    for i in list:
        key.append(i['name'])
    return key
```

```python
movies['keywords']=movies['keywords'].apply(keyword_split)
movies.head()
```

```python
movies['cast'][0][:1000]
```

```python
def cast_split(text):
    list= json.loads(text)
    cast=[]
    for i in list:
        cast.append(i['name'])
    return cast
```

```python
movies['cast']=movies['cast'].apply(cast_split)
movies.head()
```

```python
movies['crew'][0][:1000]
```

```python
def crew_split(text):
    list= json.loads(text)
    crew=[]
    for i in list:
        if i["job"]=="Director":
          crew.append(i['name'])
    return crew
```

```python
movies['crew']=movies['crew'].apply(crew_split)
movies.head()
```

```python
def remove_space(l):
    for i in range(len(l)):
        l[i]=l[i].replace(" ",'')
        l[i]=l[i].lower()
    return l
```

```python
movies['genres']=movies['genres'].apply(remove_space)
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(remove_space)
movies.head()
```

```python
movies['cast']=movies['cast'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: " ".join(x))
movies['keywords']=movies['keywords'].apply(lambda x: " ".join(x))
movies['cast']=movies['cast'].apply(lambda x: " ".join(x))
movies['crew']=movies['crew'].apply(lambda x: " ".join(x))
movies.head()
```

```python
movies['description']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies.head()
```

```python
new_movies=movies[['title','movie_id','description']]
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer=CountVectorizer(max_features=5000,stop_words="english")
x=vectorizer.fit_transform(new_movies['description']).toarray()
```

```python
x
```

```python
distance_similarity = cosine_similarity(x)
```

```python
def recommend1(name):
    movies_index = new_movies.loc[(movies["title"] == name)].index[0]
    dist_simi = list(enumerate(distance_similarity[movies_index]))

    dist_simi.sort(reverse=True, key=lambda x: x[1])

    return [i[0] for i in dist_simi[1:6]]
```

```python
recommend1('')
```

```python
def recommand(name,n=5):
    index=new_movies[new_movies["title"]==name].index
    simi=cosine_similarity(x[index],x).flatten()
    indexed_simi = list(enumerate(simi, start=0)) # combine with index
    sorted_simi = sorted(indexed_simi, key=lambda x: x[1], reverse=True)
    l=[]
    
    for i in range(1,n+1):
        j = sorted_simi[i][0]
        l.append(new_movies['title'][j])
    return l
```

```python
recommend1('Spectre')
```

```python
recommand('Spectre')
```

```python
l=[]
for i in new_movies['title']:
    l.append(recommand(i))
l
```
